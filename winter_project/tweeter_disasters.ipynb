```python
import re

def clean_text(text):
    new_text = re.sub(r'@.+?[\s\t\n\r]', '', text)
    new_text = re.sub(r'http(s)?://.+[\s\t\n]?', '', new_text)
    new_text = re.sub(r'[^\w\s]', ' ', new_text)
    new_text = re.sub(r'([A-Z])', lambda x: x.group().lower(), new_text)
    new_text = re.sub(r'\s+', ' ', new_text)
    return new_text.strip()

```

```python
def generate_barplot(count_matrix, words, num_of_bars, title, xlabel, ylabel, inverse=False):
    words_count = count_matrix.sum(axis=0)
    if inverse:
        sorted_idx = np.squeeze(np.asarray(np.argsort(words_count)[0, :num_of_bars]))
    else:
        sorted_idx = np.squeeze(np.asarray(np.argsort(words_count)[0,-num_of_bars:]))
    words = np.array(words)
    words_batch = words[sorted_idx]
    words_batch_count = np.squeeze(np.asarray(words_count[0, sorted_idx]))
    
    plt.figure(figsize=(15, 10))
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    indexes = np.arange(len(words_batch))
    width = 0.8
    plt.bar(indexes, words_batch_count, width)
    plt.xticks(indexes , words_batch, rotation=45)
    plt.show()
```

```python
from contextlib import contextmanager
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, classification_report, average_precision_score,\
plot_precision_recall_curve, plot_roc_curve
from sklearn.model_selection import StratifiedKFold, cross_val_score


class_names = ['disaster', 'fake']
dataset = tuple([X_train, y_train, X_test, y_test])


@contextmanager
def model_report(model, x_train, y_train, x_test, y_test):
    try:
        model = model.fit(x_train, y_train)
        y_pred = model.predict(X_test)
        y_pred_proba = model.predict_proba(X_test)
        print(f'Accuracy: {sum(y_pred==y_test)/len(y_test)}')
        print(classification_report(y_test, y_pred, target_names=class_names))
        
        yield {
            'model': model,
            'pred': y_pred
        }
    finally:
        disp = plot_confusion_matrix(
            model, x_test, y_test,
            display_labels=class_names,
            cmap=plt.cm.Purples,
            normalize='all'
        )
        disp.ax_.set_title(f'Confusion matrix of {model}')

        average_precision = average_precision_score(y_test, y_pred_proba[:, 1])
        disp = plot_precision_recall_curve(model, X_test, y_test)
        disp.ax_.set_title(f'{model} Precision-Recall curve: '
                   'Average precision={0:0.2f}'.format(average_precision))
        
        disp = plot_roc_curve(model, X_test, y_test)
        disp.ax_.set_title(f'{model} ROC')
        
        kf = StratifiedKFold(5, shuffle=True, random_state=1)
        auc = cross_val_score(
            model,
            X_train, 
            y_train, 
            scoring="roc_auc", 
            cv=kf
        )
        print(f'Mean auc: {auc.mean()}')
        
        plt.show()  
        
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

with model_report(model, *dataset) as report:
    model = report['model']
    y_pred = report['pred']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import itertools
import nltk

nltk.download("stopwords")
from nltk.corpus import stopwords

stop_words = stopwords.words('english')

train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

class TextCleaning(object):
    
    def __init__(self, stop_words):
        self.stop_words = set(
            x for x in itertools.chain.from_iterable(
                self.__clean_text(w).split() for w in stop_words
            )
        )
    
    @staticmethod
    def __clean_text(text):
        new_text = re.sub(r'@.+?[\s\t\n\r]', '', text)
        new_text = re.sub(r'http(s)?://.+[\s\t\n]?', '', new_text)
        new_text = re.sub(r'[^\w\s]', ' ', new_text)
        new_text = re.sub(r'([A-Z])', lambda x: x.group().lower(), new_text)
        new_text = re.sub(r'\s+', ' ', new_text)
        return new_text.strip()
    
    @staticmethod
    def __remove_stop_words(text):
        return ' '.join(x for x in text.split() if x.lower() not in stop_words)
    
    def __call__(self, df):
        df_new = df.fillna('')
        df_new = df_new.apply(self.__clean_text)
        df_new = df_new.apply(self.__remove_stop_words)
        return df_new

X_train, y_train  = train_df['text'], train_df['target']
X_test = test_df['text']


pipe = Pipeline([
    ('cleaning', FunctionTransformer(TextCleaning(stop_words))),
    ('count', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('estimator', LogisticRegression())
],
    verbose = True
)

pipe.fit(X_train, y_train)
y_test = pipe.predict(X_test)
plt.hist(y_test)

test_df['target'] = y_test
test_df = test_df[['id', 'target']]
test_df.to_csv('submission.csv', columns=['id', 'target'], index=False)
```
